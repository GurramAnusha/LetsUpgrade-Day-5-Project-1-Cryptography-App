```python
from cryptography.fernet import Fernet  # importing Fernet module from cryptography.fernet package
```

```python
def generateKey():                                 # generateKey(), this function is used to generate the key 
    key = Fernet.generate_key()                    # To generate key we are using generate_key() function in Fernet module
    print(key)                                     # printing the key generated
    print(type(key))                               # printing the type of the key
    key_file = open("PasswordKey.key",'wb')        # we are storing the key value (i.e in bytes,so open file in'wb' mode) in one file with extention .key 
    key_file.write(key)
    key_file.close()
```

```python
generateKey()
```

```python
def getMyKey():                                    # getMyKey(), this function will return the key generated by reading the key content from the file(PasswordKey.Key)
    key_content = open("PasswordKey.key",'rb')     # opening the file in 'rb' mode since the key type is bytes
    key = key_content.read()
    key_content.close()
    return key                                     # returning key value 
```

```python
getMyKey()
```

```python
def getContentFromUser():                          # getContentFromUser(), this function will take the input from user and return the content 
    return input("Enter the Content you want to Encrypt:\n")
```

```python
getContentFromUser()
```

```python
def encryptMessage(normal_message):                # encryptMessage(), this function will encrypt the normal message entered by user and return encrypted message
    key = getMyKey()
    k = Fernet(key)
    encrypted_Message = k.encrypt(normal_message)  # encrypt() will encrypt the normal message
    return encrypted_Message
```

```python
encryptMessage(b"HI")
```

```python
def decryptMessage(encrypted_message):             # decryptMessage(), this function will decrypt the encrypted message and return decrypted message
    key = getMyKey()
    k = Fernet(key)
    decrypted_Message = k.decrypt(encrypted_message) # decrypt() will decrypt the encrypted message
    return decrypted_Message
```
